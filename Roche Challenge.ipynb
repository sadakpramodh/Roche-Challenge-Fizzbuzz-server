```python
# read inputs and type conversion
int1, int2, limit, str1, str2 = input().split()

int1 = int(int1)
int2 = int(int2)
limit = int(limit)
output = list()
```

```python
print(int1, int2, limit, str1, str2)
```

```python
# logic implementation

for i in range(1, limit+1):
    if i % int1 == 0 and i % int2 == 0:
        output.append(str1 + str2)
    elif i % int1 == 0:
        output.append(str1)
    elif i % int2 == 0:
        output.append(str2)
    else:
        output.append(str(i))
```

```python
# conversion of output list to string

def list_to_string(list):
    string = ""
    for i in list:
        string += i+","
    return string[:-1]
```

```python
list_to_string(output)
```
